## 2. If-Statements with User-Defined Function

```python
# Create a User defined function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Using Defined function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check the values in price_range

df['price_range'].value_counts(dropna = False)
```

```python
# Check what the most expensive product

df['prices'].max()
```

## If-Statements with the loc() Function

```python
# First If-Statements with the loc() Function

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Second If-Statements with the loc() Function

df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
# Third If-Statements with the loc() Function

df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Value check, count conclusion
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Repeat for entire dataframe 
# First If-Statements with the loc() Function

df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, ' price_range_loc'] = 'High-range product'
```

### 5. Task - Deriving New Variables

#### 5.1 Busiest days

```python
# Busiest days column with two busiest days of the week and Slowest days with two slowest days od the week
# With If-statment and loop function

result_2 = []

for value in df_ords_prods_merged["order_dow"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Least busy days")
  else:
    result_2.append("Regularly busy days")
```

```python
# Combining results of loop function with dataframe

df_ords_prods_merged['busiest_days'] = result_2
```

```python
# Value check, count conclusion

df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

```python
# Check for the new column

df_ords_prods_merged.head()
```

The numbers in count conclusion in both columns busiest_day and busiest_days are different becouse the fit is different, in the first one we have only one day, the most intense, in the second column we have two days specified as the most intense, the same applies to the lowest days. Other criteria are equal to other results.

#### 5.2 The busiest hours of the day

```python
# Looking for the busiest hours of the day, count conclusion

df_ords_prods_merged['order_hour_of_day'].value_counts()
```

I divide the results into three equal groups. A day consists of 24 hours, divided into 3, giving 8 hours a day for each group.

```python
# Create new column with If-statment and loop function
hour_day = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hour_day.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hour_day.append("Fewest orders")
  else:
    hour_day.append("Average orders")
```

```python
# Combining results of loop function with dataframe

df_ords_prods_merged['busiest_period_of_day'] = hour_day
```

```python
# Check for the new column

df_ords_prods_merged.head()
```

```python
# Check nr.2

df_ords_prods_merged[['order_hour_of_day', 'busiest_period_of_day']].head()
```

#### 5.3 Print the frequency

```python
# Value check, count conclusion

df_ords_prods_merged['busiest_period_of_day'].value_counts()
```

The sums agree with the numbers by grouping into three groups of 8 hours
