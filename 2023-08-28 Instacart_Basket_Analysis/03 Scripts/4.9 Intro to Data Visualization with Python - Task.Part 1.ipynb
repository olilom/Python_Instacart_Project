#### 2. Data Wrangling

```python
# Columns renaming
cust.rename(columns = {'First Name' : 'name'}, inplace = True)
cust.rename(columns = {'Surnam' : 'surname'}, inplace = True)
cust.rename(columns = {'Gender' : 'gender'}, inplace = True)
cust.rename(columns = {'STATE' : 'state'}, inplace = True)
cust.rename(columns = {'Age' : 'age'}, inplace = True)
cust.rename(columns = {'n_dependants' : 'dependants_number'}, inplace = True)
```

```python
# Checking the new columns names
cust.head()
```

```python
#Reducing the data format
cust['dependants_number'] = cust['dependants_number'].astype('uint8')
cust['age'] = cust['age'].astype('uint8')
```

```python
# Output check
cust.info()
```

```python
# Finding missing Values
cust.isnull().sum()
```

```python
# Cheching the 'name' column
cust['name'].value_counts(dropna = False)
```

```python
# Rows only with missing value
# Subset is created 
cust_nan=cust[cust['name'].isnull() == True]
```

```python
cust_nan
```

```python
# Replace NaN with "Unknown"
cust['name'].fillna('Unknown', inplace=True)
```

```python
# Check for missing Values
cust.isnull().sum()
```

```python
#Cheching for mixed data type in columns
for col in cust.columns.tolist():
  weird = (cust[[col]].applymap(type) != cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (cust[weird]) > 0:
    print (col)
```
